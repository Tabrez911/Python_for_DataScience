# BUILDING DASHBOARD IN PYTHONS:

```python
#So what is a dashboard.

#How do we build a dashboard in Python.

#Well a dashboard is a combination of charts.

#figure of two plots or more than that is actually a dashboard because it combines two or more different

#We're going to be using the subplots function.

#We're going to be using the subplots function to create dashboards and populate those dashboards with different types of chart.

#So let's go ahead and do that.
```

```python
f, axes = plt.subplots(2,2, figsize=(12,12))

k2 = sns.kdeplot(movies.BudgetMillion, movies.AudienceRating, ax=axes[0,0])                      #kdeplot

k3 = sns.kdeplot(movies.BudgetMillion,movies.CriticRating, ax=axes[0,1])                         #kdeplot

k2.set(xlim=(-20,160)) 
k3.set(xlim=(-20,160)) 

v1 = sns.violinplot(x='Years', y='CriticRating', data=movies[movies.Genre=='Drama'],
                    ax=axes[1,0])                                                                #violinplot


k1 = sns.kdeplot(movies['CriticRating'], movies['AudienceRating'],                               #kdeplot
                 shade=True, shade_lowest=False, cmap='Reds')
#Tip:
k1 = sns.kdeplot(movies['CriticRating'], movies['AudienceRating'], cmap='Reds',ax=axes[1,1])
```

```python
#Lets make some changes and see:

#what if we want to add a plot that is not a seaborn plot into our diagram.

#So let's plot on hist for the critic ratings 
```

```python
f, axes = plt.subplots(2,2, figsize=(12,12))

k2 = sns.kdeplot(movies.BudgetMillion, movies.AudienceRating, ax=axes[0,0],
                 shade=True, shade_lowest=False, cmap='inferno')                      #kdeplot

k3 = sns.kdeplot(movies.BudgetMillion,movies.CriticRating, ax=axes[0,1])                         #kdeplot

k2.set(xlim=(-20,160)) 
k3.set(xlim=(-20,160)) 

v1 = sns.violinplot(x='Years', y='CriticRating', data=movies[movies.Genre=='Drama'],
                    ax=axes[1,0])                                                                #violinplot


h = plt.hist(movies.CriticRating, bins=15, ax=axes[1,1])                                         #histplot
```

```python
#The above code will give an error coz,
#for non seaborne plots the procedures are a bit different.
#Lets see and fix this error.
```

```python
f, axes = plt.subplots(2,2, figsize=(12,12))

k2 = sns.kdeplot(movies.BudgetMillion, movies.AudienceRating, ax=axes[0,0])                      #kdeplot

k3 = sns.kdeplot(movies.BudgetMillion,movies.CriticRating, ax=axes[0,1])                         #kdeplot

k2.set(xlim=(-20,160)) 
k3.set(xlim=(-20,160)) 

v1 = sns.violinplot(x='Years', y='CriticRating', data=movies[movies.Genre=='Drama'],
                    ax=axes[1,0])                                                                #violinplot


axes[1,1].hist(movies.CriticRating, bins=15)                                                     #histplot
plt.show()
```

```python
# As we can see above error has been fixed.
# We can see there's our chart.
# It has been placed in there so that is the standard pipelined approach.

# Important point to note: for seaborn there is different approach and for matplot there is different approach
```
