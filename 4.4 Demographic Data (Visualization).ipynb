```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
stats = pd.read_csv('Demographic-Data.csv')
```

```python
stats.head()
```

```python
stats.rename(columns={'Country Name':'CountryName','Country Code':'CountryCode','Birth rate':'BirthRate',\
                      'Internet users':'InternetUsers', 'Income Group': 'IncomeGroup'},inplace=True)
```

```python
stats.head()
```

```python
#Visualization

vis1 = sns.distplot(stats['InternetUsers'])
```

```python
vis2 = sns.distplot(stats['InternetUsers'], bins=30)
```

```python
vis3 = sns.boxplot(x='IncomeGroup', y='BirthRate', data=stats)
```

```python
vis4 = sns.lmplot(x='InternetUsers', y='BirthRate', data=stats)
```

```python
vis4 = sns.lmplot(x='InternetUsers', y='BirthRate', data=stats, fit_reg=False)
```

```python
vis4 = sns.lmplot(x='InternetUsers', y='BirthRate', data=stats, fit_reg=False, hue='IncomeGroup')
```

```python
vis4 = sns.lmplot(x='InternetUsers', y='BirthRate', data=stats, fit_reg=False, hue='IncomeGroup', size=10)
```

```python
vis4 = sns.lmplot(x='InternetUsers', y='BirthRate', data=stats, fit_reg=False, hue='IncomeGroup', size=10, aspect=2)
```

```python
vis5 = sns.boxenplot(x='IncomeGroup', y='BirthRate', data=stats)
```
